```python
class WaterJug:

  # The states are measured using two integers
  def __init__(self):
    self.leftJug = 0
    self.leftJugMax = 5
    self.rightJug = 0
    self.rightJugMax = 3

  # Check for Negative values and Overflows in Left Jug
  def validLeftJug(self):
    return (self.leftJug >= 0) and (self.leftJug <= self.leftJugMax)

  # Check for Negative values and Overflows in Right Jug
  def validRightJug(self):
    return (self.rightJug >= 0) and (self.rightJug <= self.rightJugMax)

  # Verify that both jugs don't have illegal states
  def stateOkay(self):
    return self.validLeftJug() and self.validRightJug()


  #
  # Rule 1: Fill the Left Jug
  #
  def fillLeftJug(self):
    self.leftJug = self.leftJugMax

  #
  # Rule 2: Fill the Right Jug
  #
  def fillRightJug(self):
    self.rightJug = self.rightJugMax

  #
  # Rule 3: Empty the Left Jug
  #
  def emptyLeftJug(self):
    self.leftJug = 0

  #
  # Rule 4: Empty the Right Jug
  #
  def emptyRightJug(self):
    self.rightJug = 0

  #
  # Rule 5: Empty the Left Jug into the Right Jug
  #
  def emptyLeftJugIntoRightJug(self):
    tempLeft, tempRight = self.leftJug, self.rightJug

    maxFill = self.rightJugMax - self.rightJug
    if maxFill < self.leftJug:
      self.rightJug = self.rightJugMax
      self.leftJug -= maxFill
    else:
      self.rightJug += self.leftJug
      self.leftJug = 0

    # Check for Illegal States
    if not self.stateOkay():
      print("Error Moving Water, Undoing State")
      self.leftJug, self.rightJug = tempLeft, tempRight



  #
  # Rule 6: Empty the Right Jug into the Left Jug
  #
  def emptyRightJugIntoLeftJug(self):
    tempLeft, tempRight = self.leftJug, self.rightJug

    maxFill = self.leftJugMax - self.leftJug
    if maxFill < self.rightJug:
      self.leftJug = self.leftJugMax
      self.rightJug -= maxFill
    else:
      self.leftJug += self.rightJug
      self.rightJug = 0

    # Check for Illegal states
    if not self.stateOkay():
      print("Error Moving Water, Undoing State")
      self.leftJug, self.rightJug = tempLeft, tempRight

  #
  # Print Current State
  #
  def printState(self):
    print(f"Left Jug: {self.leftJug}    Right Jug: {self.rightJug}")

  #
  # Check Win Condition
  #
  def successCondition(self):
    if self.leftJug == 4:
      print("Success!")
    else:
      print("Failure...")
```

```python

#Initialize the Water Jugs
wj = WaterJug()
wj.printState()
print("-----------------------")

# Fill the Left Jug First
wj.fillLeftJug()
wj.printState()
print("-----------------------")

# Move the Left Jug into the Right Jug
wj.emptyLeftJugIntoRightJug()
wj.printState()
print("-----------------------")

# Empty the Right Jug
wj.emptyRightJug()
wj.printState()
print("-----------------------")

# Move the Left Jug Into the Right Jug Again
wj.emptyLeftJugIntoRightJug()
wj.printState()
print("-----------------------")

# Completely fill the Left Jug
wj.fillLeftJug()
wj.printState()
print("-----------------------")

# Empty the Left into the Right
wj.emptyLeftJugIntoRightJug()
wj.printState()
print("-----------------------")

# Check Win Condition
wj.successCondition()

```
